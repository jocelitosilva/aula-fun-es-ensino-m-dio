```python
#função quadratica e linear 1º passo para integrar
import matplotlib.pyplot as plt #ferramenta para plotar gráfico
import numpy as np #ferramenta matemática para arrays 
from datetime import date #ferramenta para data 
```

```python
x=np.linspace(-5, 3)#range de x na equação quadratica
b=np.linspace(-3, 5)#range de b na linear
a=np.linspace(-3, 2)#renga de a na linear
y =x**2+2*x-1#equação de 2grau quadratica
reta=a*x+b#equação de 1grau linear
data_e_hora_local = (datetime.today().strftime('%d/%m/%Y %H:%M'))
```

```python
plt.title('função quadratica e linear')#plotando titulo
plt.plot(x,y, color = 'g') #definindo cor da parabola
plt.plot(a,b, color = 'b') #definindo cor da reta
plt.grid() #plotando grade
plt.show() #plote
print(data_e_hora_local)#dados de tempo
```
